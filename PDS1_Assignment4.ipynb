## Q2 (1 point)
In the ufo dataset below, find five locations, in which there were the most frequent number of ufo reports during the period before 1980 (not including this year). Show the location name and number of reports in one dataframe as an answer. Note that location is City and State together.

***Note: You will get full 1 point only in case of having correct answer. No points for trials and wrong answers.***

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head()
```

```python
ufo['Time'] = pd.to_datetime(ufo['Time'])
ufo['Year'] = pd.DatetimeIndex(ufo['Time']).year

new_ufo = ufo[ufo['Year'] < 1980]
new_ufo = new_ufo.groupby(["City", "State"]).size().reset_index(name="Frequency")

new_ufo.sort_values(by = ['Frequency'], inplace=True, ascending=False)
new_ufo.head()
```

In the orders dataset below, answer the following three tasks:

3. Find five most revenue generating items along with revenue generated - 0.5 points.

For all tasks no need to split item_name values, assume that those are the items and item pairs available in menu (i.e. there are 50 unique positions). For task three you need to create additional column. Use .head() to show the result of each question.

```python
orders = pd.read_table('http://bit.ly/chiporders')
orders.head()
```

```python
# just showing that there are 50 positions
orders.item_name.nunique()
```

```python
#Task 1
ans4 = orders.groupby(["item_name"]).size().reset_index(name="freq")
ans4.sort_values(by = ['freq'], inplace=True, ascending=False)

ans4.head()
```

```python
#Task 2
ans5 = orders.groupby(['item_name']).sum().reset_index()
ans5 = ans5.sort_values(by = ['quantity'], ascending=False)
ans5.head()
```

```python
#Task 3
ans4['item_price'] = orders['item_price'].str.replace('$', '')
ans4['item_price'] = ans4['item_price'].astype(float)

ans4['revenue'] = ans4['freq'] * ans4['item_price']
ans4.sort_values(by = ['revenue'], inplace=True, ascending=False)

ans4.head()
```

## Q4 (1 point)
In the IMBD ratings dataset below, find Top 12 actors (i.e. actors that appeared the most frequently in the movies from the dataset). 

Then create a dataframe with two columns: Full name of an actor and Number of movies he/she was listed within this dataset. Finally, show the top 12 actors with corresponding frequencies using .head().

***Note: You will get full 1 point only in case of having correct answer. No points for trials and wrong answers.***

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
movies.actors_list.replace(['\[', 'u\'','\'\]', 'u"', '\"'],'', regex=True, inplace=True)
movies.actors_list.replace('\',', ',', regex=True, inplace=True)

df = pd.DataFrame(movies.actors_list.str.split(',').tolist(), columns = ['actor_1','actor_2','actor_3'])

frames = [df['actor_1'], df['actor_2'], df['actor_3']]

result = pd.concat(frames)
result = pd.DataFrame(result).reset_index(drop = True)
result = result.rename(columns = {0:'Name'})

result['freq'] = result.groupby('Name')['Name'].transform('count')
result.sort_values(by = ['freq'], inplace=True, ascending=False)

result = result.drop_duplicates()
result.head(12)
```

## Q5 (1 point)
In the IMBD ratings dataset below, create a new column with duration categories such as: 60-89.99, 90-119.99, 120-149.99 and so on (i.e. with 30 minutes step) until you will cover the movie with the longest duration. 

Then for each duration category find a movie with the highest star_rating along with its title and actual duration (i.e. no need to show content_rating, actors_list, genre). Save it in one dataframe sorted by duration in an ascending order.

***Note: You will get full 1 point only in case of having correct answer. No points for trials and wrong answers.***

***Hint: you might need to use methods .drop() and .merge(), but it depends on your solution.***


```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
bins = [60, 89, 119, 149, 179, 209, 239, np.inf]
names = ['60-89', '90-119', '120-149', '150-179', '180-209', '210-239', '240-269']

movies['dur_cat'] = pd.cut(movies['duration'], bins, labels=names)

c1 = movies[movies['dur_cat'] == '60-89']
c2 = movies[movies['dur_cat'] == '90-119']
c3 = movies[movies['dur_cat'] == '120-149']
c4 = movies[movies['dur_cat'] == '150-179']
c5 = movies[movies['dur_cat'] == '180-209']
c6 = movies[movies['dur_cat'] == '210-239']
c7 = movies[movies['dur_cat'] == '240-269']

c1 = c1[c1.star_rating == c1.star_rating.max()]
c2 = c2[c2.star_rating == c2.star_rating.max()]
c3 = c3[c3.star_rating == c3.star_rating.max()]
c4 = c4[c4.star_rating == c4.star_rating.max()]
c5 = c5[c5.star_rating == c5.star_rating.max()]
c6 = c6[c6.star_rating == c6.star_rating.max()]
c7 = c7[c7.star_rating == c7.star_rating.max()]

frames = [c1, c2, c3, c4, c5, c6, c7]

result = pd.concat(frames)
result
```
