## Q1 (1 point)
Prove the Bias-Variance tradeoff decomposition, where MSE can be written as a sum of Variance of the estimator, Bias (squared) of the estimator and an irreducible error. 

Variance - variability of model prediction for a given value that informs us about the spreading of our data.

Bias - difference between our model's average prediction and the correct value we are attempting to calculate.

Irreducible error - measure of the amount of noise in our data.

Let's call the variable we're trying to predict Y.
$$Y=f(X) + error$$

The estimated squared error at a point x using linear regression is:

$$Err(x) = E[(Y-\hat{f}(x)^2)]$$

$$Err(x) = (E[\hat{f}(x)^2)] - f(x))^2 + E[(\hat{f}(x)-E[\hat{f}(x)])^2] + \sigma_e^2$$

$$Err(x) = Bias^2 + Variance + IrreducibleError$$

If the model is too simple and has small amount of parameters then it may have low variance and high bias. So we need to find good balance without overfitting and underfitting the data.

This tradeoff in complexity is why there is a tradeoff between bias and variance. In other words, more complicated and less complex algorithm cannot simultaneously occur. For this reason it is important to consider prediction models' behavior and variance.

- read netflix data
- find ratios of Movies and TV Shows for different countries
- plot histogram of average duration length for movies with following genres: "dramas", "comedies", "sports movies" (for this case use only first genre)
- create pivot table with count of word "love" in description for different genres (for this case use only first genre)

```python
# Task 1
import numpy as np
import collections as c

netflix = pd.read_csv("netflix_titles.csv")
netflix
```

```python
#Task 2

new = netflix['country'].str.split(', ', expand = True)
new = new.fillna(value=np.nan)

d = new.values.tolist()
flat_list = list(np.concatenate(d).flat)

fr = c.Counter(flat_list)

fr1 = dict(sorted(fr.items(), key=lambda x: x[1]))

li2 = netflix['type'].value_counts()
overall_ratio = li2['Movie']/li2['TV Show']

num = list(fr1.values())
overall_ratio
```

```python
calcyli = []
calcxli = []

def calcy(num, overall_ratio, calcyli):
    
    for i in num:
        y = i / (overall_ratio + 1)
        calcyli.append(y)
        
    return(calcyli)

def calcx(y, overall_ratio, calcxli):
    
    for i in y:
        
        x = i * overall_ratio
        calcxli.append(x)
        
    return(calcxli)

yto = calcy(num, overall_ratio, calcyli)
xto = calcx(yto, overall_ratio, calcxli)
```

```python
each_cora = list(zip(xto, yto))
rrrr = [(k,v) for k,v in fr1.items()]
cora = list(zip(each_cora, rrrr))
cora
```

```python
# Task 1
df = pd.read_csv("knn_bias_variance_tradeoff.csv")
df.head()
```

```python
# Task 2
from sklearn.model_selection import train_test_split

columns = ['XVPM', 'GWYH', 'TRAT', 'TLLZ', 'IGGA', 'HYKR', 'EDFS', 'GUUB', 'MGJM', 'JHZC']

x = df[columns]
y = df['TARGET CLASS']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
# Task 3
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
data_df = ss.fit_transform(df)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor

train_score = []
test_score = []
k_vals = []

for k in range(1, 50):
    k_vals.append(k)
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(x_train, y_train)
    
    tr_score = knn.score(x_train, y_train)
    train_score.append(tr_score)
    
    te_score = knn.score(x_test, y_test)
    test_score.append(te_score)
```

```python
plt.figure(figsize=(10,5))
plt.xlabel('Different Values of K')
plt.ylabel('Model score')
plt.plot(k_vals, train_score, color = 'r', label = "training score")
plt.plot(k_vals, test_score, color = 'b', label = 'test score')
plt.legend(bbox_to_anchor=(1, 1),
           bbox_transform=plt.gcf().transFigure)
plt.show()
```
